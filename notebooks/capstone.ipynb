### MNPS High School Demographics

```python
mnps_hs_dem = mnps_hs[['School Level', 'School ID',\
                            'School Name', 'State School ID',\
                            'Zip Code', 'Total Enrollment',\
                            'Economically Disadvantaged',\
                            'Students with Disabilities',\
                            'Limited English Proficiency',\
                            'Female', 'Male',\
                            'White', 'Black or African American',\
                            'Hispanic/Latino', 'Asian',\
                            'American Indian or Alaska Native',\
                            'Native Hawaiian or Other Pacific Islander',\
                            'DataAsOf']]
```

```python
# mnps_hs_dem
```

```python
# mnps_hs_dem.loc[4,'Economically Disadvantaged'].replace('%','ZZZ')
```

```python
# df['result'] = df['result'].str.replace(r'\D', '')
```

```python
mnps_hs_dem['econ_dis_pct'] = mnps_hs_dem['Economically Disadvantaged']\
    .str.replace(r'\D','').astype(float)
```

```python
mnps_hs_dem.info()
```

```python
mnps_hs_dem.head(2)
```

```python
mnps_hs_dem['econ_dis_decimal'] = mnps_hs_dem['econ_dis_pct']
```

```python
mnps_hs_dem.head(2)
```

```python
mnps_hs_dem['econ_dis_decimal']/= 100.
```

```python
mnps_hs_dem.head(2)
```

```python
mnps_hs_dem['econ_dis_count'] = mnps_hs_dem['Total Enrollment']*mnps_hs_dem['econ_dis_decimal']
```

```python
mnps_hs_dem.head(2)
```

```python
mnps_hs_dem_sorted = mnps_hs_dem.sort_values(by = 'econ_dis_pct', ascending = False)
```

```python
mnps_hs_dem_sorted.head()
```

```python
plt.bar(mnps_hs_dem_sorted['School Name'],\
        mnps_hs_dem_sorted['econ_dis_count'],\
        color = '#53b92b')
plt.xticks(rotation=90)
plt.title('Count of Economically Disadvantaged Students per High School')
plt.xlabel('High School')
plt.ylabel('Count of Students')
plt.xticks(rotation=45,
        horizontalalignment="right")
# plt.ylim(0, 3250) 
;
```

```python
plt.bar(mnps_hs_dem_sorted['School Name'],\
        (mnps_hs_dem_sorted['econ_dis_count']/mnps_hs_dem_sorted['Total Enrollment']*100),\
        color = '#53b92b')
plt.xticks(rotation=90)
plt.title('Calculated Percentage of Economically Disadvantaged Students per High School')
plt.xlabel('High School')
plt.ylabel('Percent of Students')
plt.xticks(rotation=45,
        horizontalalignment="right")
# plt.ylim(0, 3250) 
;
```

```python
plt.bar(mnps_hs_dem_sorted['School Name'],\
        mnps_hs_dem_sorted['econ_dis_pct'],\
        color = '#53b92b')
plt.xticks(rotation=90)
plt.title('Percentage of Economically Disadvantaged Students per High School')
plt.xlabel('High School')
plt.ylabel('Percent of Students')
plt.xticks(rotation=45,
        horizontalalignment="right")
# plt.ylim(0, 3250) 
;
```
